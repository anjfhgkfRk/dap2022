## 2-1 확인문제


### 04. 다음 함수의 그래프를 그리세요.
(a) $y=3x+1$<br>
(b) $y=-2x^2+4x+1$

```python
import numpy as np
import matplotlib.pyplot as plt

#(a) y = 3x + 1
x = np.linspace(0, 10, 501)
y = 3*x + 1

plt.plot(x, y)
plt.show()

__answer_1 = y
```

```python
import numpy as np
import matplotlib.pyplot as plt

#(b) y = -2x^2 + 4x + 1
x = np.linspace(-10, 10, 501)
y = -2*x**2 + 4*x + 1

plt.plot(x, y)
plt.show()

__answer_2 = y
```

## 2-2 확인문제

### 05. 다음 함수의 그래프를 그리세요.
(a) $f(x)=-2^{x-1}+3$<br>
(b) $g(x)=\log_{0.5}(-x+1)-2$

```python
import numpy as np
import matplotlib.pyplot as plt

#(a) f(x) = -2^(x-1)+3
x = np.linspace(0, 10, 501)
y = -2**(x - 1) + 3

plt.plot(x, y)
plt.show()

__answer_3 = y
```

```python
import numpy as np
import matplotlib.pyplot as plt

#(b) g(x) = log_0.5(-x+1)-2
x = np.linspace(-10, 0.9, 501)
y = np.log2(1/(-x+1)) -2

plt.plot(x, y)
plt.show()

__answer_4 = y
```

## 2-3 확인문제


### 05. 다음 함수의 그래프를 그리세요.
(a) $f(x)=\sin{3(x-1)}+1$<br>
(b) $g(x)=\tan{2x}-1$

## 3-1 확인문제

### 04. 다음의 합을 **반복문**을 이용하여 구해서 변수 **sum**에 저장하세요.
(a) $\sum^{10}_{k=1}{(2k^3-k^2+1)}$<br>
(b) $\sum^{10}_{k=1}{(2^{k-1}+1)}$


```python
#(a)
sum = 0
for i in range(1, 11):
  sum += 2*(i**3) - i**2 + 1

print(sum)

__answer_7 = sum
```

```python
#(b)
sum = 0
for k in range(1, 11):
  sum += 2**(k-1) +1

print(sum)

__answer_8 = sum
```

## 3-2 확인문제

### 03. 각 구간에서 다음 일반항의 그래프를 그리세요.<br>
(a) $a_n={{2^n-3^n}\over{5^n+6^n}}$<br>
(b) $b_n = {{4^n+2^n}\over{3^{n+1}}}$


```python
import numpy as np
import matplotlib.pyplot as plt

#(a)
x = np.linspace(0, 20, 501)
y = (2**x - 3**x)/(5**x + 6**x)

plt.plot(x, y)
plt.show()

__answer_9 = y
```

```python
import numpy as np
import matplotlib.pyplot as plt

#(b)
x = np.linspace(80, 100, 501)
y = (4**x + 2**x)/(3**(x+1))

plt.plot(x, y)
plt.show()

__answer_10 = y
```
